```python
import pandas as pd

# Load the CSV file
file_path = "dummy_bms_log.csv"  # Ensure correct file path
bms_data = pd.read_csv(file_path)

# Convert timestamp column to datetime format
bms_data['Timestamp'] = pd.to_datetime(bms_data['Timestamp'])

# Fill missing values only for numeric columns
numeric_columns = ["Voltage (V)", "Current (A)", "Temperature (째C)", "SOC (%)", "SOH (%)"]
bms_data[numeric_columns] = bms_data[numeric_columns].fillna(bms_data[numeric_columns].mean())

# Display cleaned data
print(bms_data.head())
```

```python
print(bms_data.describe())

# Check SOC & SOH trends
print("Average SOC:", bms_data["SOC (%)"].mean())
print("Average SOH:", bms_data["SOH (%)"].mean())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(bms_data["Timestamp"], bms_data["Voltage (V)"], label="Voltage (V)", color="blue")
plt.xlabel("Time")
plt.ylabel("Voltage (V)")
plt.title("Battery Voltage Over Time")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
